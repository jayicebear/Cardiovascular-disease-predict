#Download data#

```python
import pandas as pd
from pandas import DataFrame, Series
from datetime import datetime
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
app = pd.read_csv('/content/drive/MyDrive/2022 fall DS440/data/APPL/AAPL_monthly_valuation_measures.csv')
JNJ = pd.read_csv('/content/drive/MyDrive/2022 fall DS440/data/JNJ/JNJ_monthly_valuation_measures.csv')
KO = pd.read_csv('/content/drive/MyDrive/2022 fall DS440/data/KO/KO_monthly_valuation_measures.csv')
META = pd.read_csv('/content/drive/MyDrive/2022 fall DS440/data/META/META_monthly_valuation_measures.csv')
NVDA = pd.read_csv('/content/drive/MyDrive/2022 fall DS440/data/NVDA/NVDA_monthly_valuation_measures.csv')
```

#Preprocessing#

```python
app = app.transpose()
app.rename(columns=app.iloc[0],inplace=True)
app = app.iloc[2:,:3]
app['MarketCap'] = app['MarketCap'].str.replace(',', '').astype(float)
app['EnterpriseValue'] = app['EnterpriseValue'].str.replace(',', '').astype(float)
app = app.apply(pd.to_numeric)
```

```python
JNJ = JNJ.transpose()
JNJ.rename(columns=JNJ.iloc[0],inplace=True)
JNJ = JNJ.iloc[2:,:3]
JNJ['MarketCap'] = JNJ['MarketCap'].str.replace(',', '').astype(float)
JNJ['EnterpriseValue'] = JNJ['EnterpriseValue'].str.replace(',', '').astype(float)
JNJ = JNJ.apply(pd.to_numeric)
```

```python
KO = KO.transpose()
KO.rename(columns=KO.iloc[0],inplace=True)
KO = KO.iloc[2:,:3]
KO['MarketCap'] = KO['MarketCap'].str.replace(',', '').astype(float)
KO['EnterpriseValue'] = KO['EnterpriseValue'].str.replace(',', '').astype(float)
KO = KO.apply(pd.to_numeric)
```

```python
META = META.transpose()
META.rename(columns=META.iloc[0],inplace=True)
META = META.iloc[2:,:3]
META['MarketCap'] = META['MarketCap'].str.replace(',', '').astype(float)
META['EnterpriseValue'] = META['EnterpriseValue'].str.replace(',', '').astype(float)
META['PeRatio'] = META['PeRatio'].str.replace(',', '').astype(float)
META = META.apply(pd.to_numeric)
```

```python
NVDA = NVDA.transpose()
NVDA.rename(columns=NVDA.iloc[0],inplace=True)
NVDA = NVDA.iloc[2:,:3]
NVDA['MarketCap'] = NVDA['MarketCap'].str.replace(',', '').astype(float)
NVDA['EnterpriseValue'] = NVDA['EnterpriseValue'].str.replace(',', '').astype(float)
NVDA = NVDA.apply(pd.to_numeric)
NVDA
```

#Fill the dates#

```python
app2012 = app2012.reset_index().rename(columns={"index": "Date"})
JNJ2012 = JNJ2012.reset_index().rename(columns={"index": "Date"})
KO2012 = KO2012.reset_index().rename(columns={"index": "Date"})
META2012 = META2012.reset_index().rename(columns={"index": "Date"})
NVDA2012 = NVDA2012.reset_index().rename(columns={"index": "Date"})
```

```python
app2012['Date'] = pd.to_datetime(app2012['Date'])
app2012 = app2012.set_index('Date')    # Changing index for Filling values

app2012 = app2012.resample('D').first()    # Resampling & Assigning values
app2012 = app2012.bfill().ffill()

JNJ2012['Date'] = pd.to_datetime(JNJ2012['Date'])
JNJ2012 = JNJ2012.set_index('Date')    # Changing index for Filling values

JNJ2012 = JNJ2012.resample('D').first()    # Resampling & Assigning values
JNJ2012 = JNJ2012.bfill().ffill()

KO2012['Date'] = pd.to_datetime(KO2012['Date'])
KO2012 = KO2012.set_index('Date')    # Changing index for Filling values

KO2012 = KO2012.resample('D').first()    # Resampling & Assigning values
KO2012 = KO2012.bfill().ffill()

META2012['Date'] = pd.to_datetime(META2012['Date'])
META2012 = META2012.set_index('Date')    # Changing index for Filling values

META2012 = META2012.resample('D').first()    # Resampling & Assigning values
META2012 = META2012.bfill().ffill()

NVDA2012['Date'] = pd.to_datetime(NVDA2012['Date'])
NVDA2012 = NVDA2012.set_index('Date')    # Changing index for Filling values

NVDA2012 = NVDA2012.resample('D').first()    # Resampling & Assigning values
NVDA2012 = NVDA2012.bfill().ffill()
```
